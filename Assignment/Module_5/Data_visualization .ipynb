```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# sample dataset for not changing the value everytime 
np.random.seed(42) 

# random data for the dataset
data = {
    'age': np.random.randint(20, 60, 100),  
    'salary': np.random.randint(30000, 100000, 100),  
    'years_experience': np.random.randint(1, 40, 100),  
    'education_level': np.random.choice(['High School', 'Associate', 'Bachelor', 'Master', 'PhD'], 100)  
}

df = pd.DataFrame(data)
df
# or you can use this for existing dataframe 
# df = pd.read_csv("abc.csv")
```

```python
# 1. Histogram for numerical distribution (for 'age' and 'salary')
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['age'], bins=15, kde=True, color='skyblue', edgecolor='black')
plt.title('Age Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['salary'], bins=15, kde=True, color='lightgreen', edgecolor='black')
plt.title('Salary Distribution')

plt.tight_layout()
plt.show()
```

```python
# 2. Scatter Plot to show correlation between 'age' and 'salary'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='salary', hue='education_level', palette='Set1')
plt.title('Age vs Salary Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# 3. Heatmap for correlation between numerical columns
# Calculate the correlation matrix
correlation_matrix = df[['age', 'salary', 'years_experience']].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
