```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Create a sample dataset
np.random.seed(42)  # For reproducibility
data = np.random.randint(20, 60, 100)  # 100 random integers between 20 and 60

# Create a Pandas DataFrame from the NumPy array
df = pd.DataFrame(data, columns=['age'])
df
```

```python
# 1. Compute Mean, Median, Mode, Variance, and Standard Deviation
mean = np.mean(df['age'])
median = np.median(df['age'])
mode = stats.mode(df['age'])[0][0] 
variance = np.var(df['age'])
std_dev = np.std(df['age'])

# Print the computed values
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

```python
# 2. Generate Histogram
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['age'], bins=15, kde=True, color='skyblue', edgecolor='black')
plt.title('Age Distribution - Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')

# 3. Generate Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='age', color='lightgreen')
plt.title('Age Distribution - Boxplot')
plt.xlabel('Age')

# Display both plots
plt.tight_layout()
plt.show()
```
