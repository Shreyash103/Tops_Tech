```python
class QuizGame:
    def __init__(self):
        # Initialize questions and answers
        self.questions = [
            {
                "question": "What is the capital of France?",
                "options": ["A) Paris", "B) London", "C) Berlin", "D) Madrid"],
                "answer": "A"
            },
            {
                "question": "Which planet is known as the Red Planet?",
                "options": ["A) Earth", "B) Mars", "C) Jupiter", "D) Saturn"],
                "answer": "B"
            },
            {
                "question": "Who wrote 'Hamlet'?",
                "options": ["A) William Shakespeare", "B) Charles Dickens", "C) J.K. Rowling", "D) Mark Twain"],
                "answer": "A"
            },
            {
                "question": "What is the largest mammal on Earth?",
                "options": ["A) Elephant", "B) Blue Whale", "C) Giraffe", "D) Rhino"],
                "answer": "B"
            },
            {
                "question": "In which year did the Titanic sink?",
                "options": ["A) 1912", "B) 1905", "C) 1923", "D) 1898"],
                "answer": "A"
            }
        ]
        self.score = 0

    def ask_question(self, question_data):
        # Ask a question and check the user's answer
        print(question_data['question'])
        for option in question_data['options']:
            print(option)
        
        # Get user's answer
        user_answer = input("Please choose the correct answer (A/B/C/D): ").upper()

        # Check if the answer is correct
        if user_answer == question_data['answer']:
            print("Correct!\n")
            self.score += 1
        else:
            print(f"Wrong! The correct answer was {question_data['answer']}. {question_data['options'][ord(question_data['answer']) - 65].split(')')[1]}\n")

    def start_game(self):
        # Start the quiz game and ask all questions
        print("Welcome to the Quiz Game!\n")
        for question_data in self.questions:
            self.ask_question(question_data)

        print(f"\nQuiz Over! Your score is: {self.score}/{len(self.questions)}")


if __name__ == "__main__":
    quiz_game = QuizGame()
    quiz_game.start_game()
```
